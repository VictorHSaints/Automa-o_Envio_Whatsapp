```python

# Fun√ß√£o para encontrar a posi√ß√£o de um elemento na tela usando uma imagem
def find_image(image_path):
    # Ajuste o n√≠vel de confian√ßa (0.8 a 1.0)
    position = pyautogui.locateOnScreen(image_path, confidence=0.8)
    if position is None:
        print(f"Imagem {image_path} n√£o encontrada.")
        return None
    return pyautogui.center(position)

# Lista de n√∫meros de telefone
numeros = []

cod = []


# Mensagem padr√£o com caracteres especiais
mensagem = """Ol√°! Obrigado por sua participa√ß√£o no CONAZ 2024 üôå

Esse √© o seu resultado com comparativos de todos os alunos que fizeram a prova em sua s√©rie. Esperamos ter voc√™ estudando no AZ Floripa em 2025 üíõüêù
"""

# Caminho da imagem do bot√£o de nova mensagem
botao_nova_mensagem = "botao_nova_mensagem.png"
botao_nova_mensagem2 = "botao_nova_mensagem2.png"

barra_mensagem = "mensagem.png"
anexo = "anexo.png"
anexo2 = "anexo2.png"

barra_local = "barra_local.png"
i = 0

# ... (Adicionar outros caminhos de imagem para os demais elementos)
time.sleep(2)

# Loop para cada n√∫mero
for numero in numeros:
    # Encontrar e clicar no bot√£o de nova mensagem
    posicao = find_image(botao_nova_mensagem)
    if posicao:
        pyautogui.click(posicao)
        time.sleep(0.5)
    posicao2 = find_image(botao_nova_mensagem2)
    if posicao2:
        pyautogui.click(posicao2)
        time.sleep(0.5)
    # Digitar o n√∫mero
    pyautogui.write(numero)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(0.5)
    
    posicao3 = find_image(barra_mensagem)
    if posicao3:
        pyautogui.click(posicao3)
        time.sleep(1)
    
    # Copiar a mensagem para a √°rea de transfer√™ncia e col√°-la
    pyperclip.copy(mensagem)
    pyautogui.hotkey("ctrl", "v")
    time.sleep(1)
    
    posicao4 = find_image(anexo)
    if posicao4:
        pyautogui.click(posicao4)
        time.sleep(0.5)
        
    posicao5 = find_image(anexo2)
    if posicao5:
        pyautogui.click(posicao5)
        time.sleep(1)
        
    posicao6 = find_image(barra_local)
    if posicao6:
        pyautogui.click(posicao6)
        time.sleep(0.5)
    
    pyautogui.write(cod[i])
    time.sleep(0.5)
    i += 1
    pyautogui.press('enter')
    time.sleep(1)
    
    # Enviar a mensagem
    pyautogui.press('enter')
    time.sleep(1)  # Ajustar o tempo de espera conforme necess√°rio
```
